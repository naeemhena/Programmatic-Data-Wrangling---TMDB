```python
# Lets identify the amount of nulls and put them into a data frame 
def null_vals(dataframe):
    '''function to show both number of nulls and the percentage of nulls in the whole column'''
    null_vals = dataframe.isnull().sum() ## How many nulls in each column
    total_cnt = len(dataframe) ## Total entries in the dataframe
    null_vals = pd.DataFrame(null_vals,columns=['null']) ## Put the number of nulls in a single dataframe
    null_vals['percent'] = round((null_vals['null']/total_cnt)*100,3) ## Round how many nulls are there, as %, of the df
    
    return null_vals.sort_values('percent', ascending=False)
```

## 4. Data Wrangling

```python
import ast
# Function to safely extract a list of country names from the JSON string
# This handles the messy format in the 'production_countries' column
def extract_names(json_str):
    try:
        # Convert the JSON string to a Python object
        data = ast.literal_eval(json_str)
        # Check if the data is a list
        if isinstance(data, list):
            # Extract the 'name' field from each dictionary in the list
            return [d.get('name') for d in data if 'name' in d]
        else:
            return None
    except (ValueError, SyntaxError):
        # Return None if the string can't be parsed
        return None

# Apply the function to create new columns with extracted genre and country names
tmdb['genre_list'] = tmdb['genres'].apply(extract_names)
tmdb['country_list'] = tmdb['production_countries'].apply(extract_names)

# Extract the release year from the release_date column and convert to Int64 type
tmdb['release_year'] = pd.to_datetime(tmdb['release_date'], errors='coerce').dt.year.astype('Int64')

# Extract the first genre name from the list (if available)
tmdb['genre_name'] = tmdb['genre_list'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
# Extract the first country name from the list (if available)
tmdb['country_name'] = tmdb['country_list'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)

# Display the first few rows of the selected columns to verify the transformations
tmdb[['title', 'release_year' ,'genre_name', 'country_name']].head()
```

```python
# Calculate ROI (Return on Investment)
# ROI = revenue / budget
tmdb['ROI'] = tmdb['revenue'] / tmdb['budget']

# Replace infinite or undefined values (e.g. where budget is 0)
tmdb['ROI'].replace([float('inf'), -float('inf')], pd.NA, inplace=True)

# Round for easier readability
tmdb['ROI'] = tmdb['ROI'].round(2)

# Quick check
tmdb[['title', 'budget', 'revenue', 'ROI']].head()
```

## 5. Exploratory Data Analysis (EDA)


### Budget vs Revenue

```python
# Calculate Return on Investment (ROI) as revenue divided by budget
tmdb['ROI'] = tmdb['revenue'] / tmdb['budget']
# Filter out movies with zero or negative budget/revenue to avoid division errors and invalid data points
tmdb = tmdb[(tmdb['budget'] > 0) & (tmdb['revenue'] > 0)]  # avoid division errors

# Create a scatter plot to visualize the relationship between budget and revenue
plt.figure(figsize=(8,6))  # Set figure size to 8x6 inches
sns.scatterplot(
    data=tmdb,           # Use the tmdb dataframe
    x='budget',          # Budget on x-axis
    y='revenue',         # Revenue on y-axis
    hue='ROI',           # Color points by ROI value
    alpha=0.6,           # Set transparency to 0.6 to handle overlapping points
    palette='viridis'    # Use viridis color palette for better visualization of ROI values
)

# Set logarithmic scales for both axes to better visualize the wide range of values
plt.xscale('log')        # Log scale for budget axis
plt.yscale('log')        # Log scale for revenue axis
plt.title('Does Spending More Make More? â€” Budget vs Revenue', fontsize=14, weight='bold')  # Add descriptive title
plt.xlabel('Budget (log scale)', fontsize=12)  # Label x-axis
plt.ylabel('Revenue (log scale)', fontsize=12)  # Label y-axis
plt.legend(title='ROI', bbox_to_anchor=(1.05, 1))  # Position legend outside the plot for clarity
plt.tight_layout()  # Adjust subplot parameters for better layout
plt.show()  # Display the plot
```

-Key insight: Strong positive correlation on log scale showing higher budgets enable higher revenue potential, overall. 
 Highest ROI in mid-range budgets

### Top 10 countries to release most movies

```python
# Explode and clean the DataFrame
tmdb_countries = tmdb.explode('country_name')  # Split rows with multiple countries into separate rows
tmdb_countries = tmdb_countries[tmdb_countries['country_name'].str.len() > 0]  # Remove rows with empty country names
tmdb_countries.dropna(subset=['release_year'], inplace=True)  # Remove rows with missing release years

# Quick check of the financial data
tmdb[['title', 'budget', 'revenue', 'ROI']].head()

# --- Prepare Data for Visualization (Top N Countries) ---

# Identify the top 10 countries by movie count
top_countries = tmdb_countries['country_name'].value_counts().head(10).index.tolist()

# Filter the dataset to include only movies from the top 10 countries
tmdb_top_countries = tmdb_countries[tmdb_countries['country_name'].isin(top_countries)]

# Group by year and country, and count the number of movies for each combination
tmdb_plot = tmdb_top_countries.groupby(['release_year', 'country_name']).size().reset_index(name='movie_count')

#  **FIX: Ensure numeric columns are standard types for plotting** ---
# Explicitly cast to standard 'int' to fix the TypeError with Seaborn.
tmdb_plot['release_year'] = tmdb_plot['release_year'].astype(int)  # Convert release_year to integer type
tmdb_plot['movie_count'] = tmdb_plot['movie_count'].astype(int)  # Convert movie_count to integer type

# Create the Static Visualization (Seaborn/Matplotlib) ---

sns.set_style("whitegrid")  # Set the visual style of the plot
plt.figure(figsize=(16, 7))  # Create a new figure with specified dimensions

# Create a line plot showing movie production trends over time for each country
sns.lineplot(
    data=tmdb_plot,
    x='release_year',
    y='movie_count',
    hue='country_name',  # Use different colors for each country
    linewidth=2.5  # Set the thickness of the lines
)

# Add title and labels to the plot
plt.title('Top 10 Countries: Movie Production Trend Over Time', fontsize=18, fontweight='bold')
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('Number of Movies Released', fontsize=14)
plt.legend(title='Production Country', bbox_to_anchor=(1.05, 1), loc='upper left')  # Position the legend outside the plot
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust the layout to make room for the legend
```

- Key insight: US dominates volume,but global competition is growing with upward trends from multiple countries. 
Germany, UK, France form strong second tier showing increasingly globalized film production landscape.

### Top countries who made the most profit

```python
# Calculate ROI (Per Movie) ---

# Calculate Return on Investment (ROI) for each movie
# Formula: (revenue - budget) / budget
# If budget is 0 or negative, set ROI to 0 to avoid division by zero or negative values
tmdb_countries['ROI'] = np.where(
    tmdb_countries['budget'] > 0,
    (tmdb_countries['revenue'] - tmdb_countries['budget']) / tmdb_countries['budget'],
    0
)
# Aggregate Data (Median ROI per Country) ---

# Calculate the median ROI for each country and sort in descending order
# Using median instead of mean to reduce the impact of outliers
avg_ROI = tmdb_countries.groupby('country_name')['ROI'].median().sort_values(ascending=False).reset_index()

# Filter to only include the top 10 countries by movie volume
# This ensures we're analyzing countries with significant sample sizes
top_10_countries_by_volume = tmdb_countries['country_name'].value_counts().head(10).index
avg_ROI_top_10 = avg_ROI[avg_ROI['country_name'].isin(top_10_countries_by_volume)].sort_values(by='ROI', ascending=False).copy()

# Create the Visualization (Bar Chart) ---

# Set up the figure size for better visualization
plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")  # Add grid lines for easier value reading

# Create a bar plot showing median ROI by country
# The magma color palette provides good contrast and visual appeal
ax = sns.barplot(
    data=avg_ROI_top_10,
    x='country_name',
    y='ROI',
    palette="magma"  # Use a visually appealing color palette
)

# Add descriptive labels and title to the plot
plt.title('Median Return on Investment (ROI) by Top 10 Production Country', fontsize=16, fontweight='bold')
plt.xlabel('Production Country', fontsize=12)
plt.ylabel('Median ROI (Return on Investment)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate country names for better readability
plt.tight_layout()  # Adjust layout to prevent label cutoff

# Add ROI values as text labels above each bar
# This makes the exact values immediately visible to viewers
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',  # Format to 2 decimal places
               (p.get_x() + p.get_width() / 2., p.get_height()),  # Position at top of bar
               ha = 'center', va = 'center',  # Center text horizontally and vertically
               xytext = (0, 9),  # Offset text slightly above bar
               textcoords = 'offset points',
               fontsize=9,
               fontweight='bold')

# Save the visualization as a PNG file for later use
plt.savefig("median_roi_by_country_bar_chart.png")

# Display the data used for the visualization
print("DataFrame used for visualization (Median ROI):")
print(avg_ROI_top_10)
```

- Key insight: Japan & India leads in median ROI though having much less movie production and budgets than USA but still lead to higher revenues hence successful movies

### Correlation between runtime and popularity

```python
correlation = tmdb['runtime'].corr(tmdb['popularity'])
print(f"Correlation between runtime and popularity: {correlation:.3f}")
```

```python
# Create a figure with specified size (9x6 inches)
plt.figure(figsize=(9,6))

# Create a scatter plot with multiple dimensions:
# - x-axis: movie runtime in minutes
# - y-axis: popularity score
# - bubble size: vote average (rating)
# - color: also representing vote average (same as size)
sns.scatterplot(
    data=tmdb,                # The DataFrame containing our movie data
    x='runtime',              # x-axis: movie duration
    y='popularity',           # y-axis: popularity metric
    size='vote_average',      # Bubble size based on rating
    hue='vote_average',       # Color based on rating
    alpha=0.6,                # Transparency of points (0.6 = 60% opaque)
    palette='magma',          # Color palette for the hue dimension
    sizes=(40, 400)           # Range of point sizes from min to max
)

# Add a title and axis labels with custom formatting
plt.title('Popularity vs Runtime (Bubble Size = Vote Average)', fontsize=14, weight='bold')
plt.xlabel('Runtime (minutes)', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)

# Position the legend outside the plot for better visibility
plt.legend(title='Vote Avg', bbox_to_anchor=(1.05, 1))

# Adjust layout to ensure everything fits properly
plt.tight_layout()

# Display the plot
plt.show()
```

-Key insight : 80-120 mins run time achieve the highest Popularity score

```python
# Define bins for runtime (you can tweak these as needed)
bins = [0, 80, 120, 150, 200, 1000]
labels = ['Short (<80)', 'Medium (80â€“120)', 'Standard (120â€“150)', 'Long (150â€“200)', 'Too long (>200)']

# Create a new column with runtime categories
# pd.cut() creates categorical bins from continuous data
# - bins: the boundaries for each category
# - labels: names for each category
# - include_lowest: ensures the lowest value is included in the first bin

tmdb['runtime_bucket'] = pd.cut(tmdb['runtime'], bins=bins, labels=labels, include_lowest=True)
```

```python
##Visualize runtime bucket vs popularity

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxenplot(
    data=tmdb,
    x='runtime_bucket',
    y='vote_average',
    palette='magma'
)

plt.title('Popularity Distribution (Average votes) by Runtime Bucket', fontsize=14, weight='bold')
plt.xlabel('Runtime Category', fontsize=12)
plt.ylabel('Vote_average', fontsize=12)
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
- Key insight: Medium-length films (80-120 mins) achieve the highest average ratings,suggesting an optimal balance for audience satisfaction.
 Ratings decline as films get longer,** with "Too long" (>200 mins) category receiving the lowest average scores.
```
